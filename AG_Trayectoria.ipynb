```python
# Mateo Velasquez Penagos
#Si no encuentra una opcion optima de fitness 16 que es el esperado, aumetnar las iteraciones o correr varias veces el algoritmo hasta que lo encunetr
import random
random.seed(42) #Si quitamos el 42 de la funcion seed podemos ver inicializaciones aleatorias diferentes
characters = ['U', 'D', 'L', 'R']
cromosomaSize = 15
population_size = 100
generations = 1000
mutation_rate = 0.02
```

```python
def generate_cromosoma(cromosomaSize):
  cromosoma = ''
  for i in range(cromosomaSize):
    cromosoma += random.choice(characters)
  return ''.join(cromosoma)
```

```python
def fitness(cromosoma, posicion=(0, 0)):
    filas, columnas = posicion
    visited = set()
    visited.add((filas, columnas))
    moveAllowed = 0

    for move in cromosoma:
        if move == 'U' and filas > 0:
            filas -= 1

        elif move == 'D' and filas < 3:
            filas += 1

        elif move == 'L' and columnas > 0:
            columnas -= 1

        elif move == 'R' and columnas < 3:
            columnas += 1

        else:
            continue

        if (filas, columnas) not in visited:
            visited.add((filas, columnas))
            moveAllowed += 1  # Recompensar por visitar una nueva casilla
        else:
            moveAllowed += 0.5  # Penalizar por revisitar una casilla

    # Penalizar si no todas las casillas fueron visitadas
    noVisited = 16 - len(visited)
    fitness = moveAllowed + noVisited*10
    return fitness
```

```python
#Seleccion metodo elitista
def selection(population):
  sorted_population = sorted(population, key=fitness, reverse=False)
  return sorted_population[0],sorted_population[1]

#Seleccion metodo por torneo
def tournament_selection(population,tournament_size):
  tournament = random.sample(population, tournament_size)
  best_in_tournament = min(tournament, key=fitness)
  best_in_tournament2 = min(tournament, key=fitness)
  return best_in_tournament, best_in_tournament2
```

```python
#Cruce en un punto
def crossover(parent1, parent2):
  point = random.randint(1, len(parent1)) # Encunetra un punto aleatorio para cruzar los cromosomas en puntos aleatorios luego en el return los concatena
  return parent1[:point] + parent2[point:]

#Cruce en dos puntos
def crossover2(parent1, parent2):
  point1 = random.randint(1, len(parent1))
  point2 = random.randint(1, len(parent1))
  if point1 > point2:
    point1, point2 = point2, point1
  return parent1[:point1] + parent2[point1:point2] + parent1[point2:]
```

```python
#Metodo de mutacion aleatoria
def mutation(cromosoma):
  cromosoma = list(cromosoma)
  for i in range(len(cromosoma)):
    if random.random() < mutation_rate:
      cromosoma[i] = random.choice(characters)
  return ''.join(cromosoma)

#Metodo de mutacion de inversion de genes
def inversionMutation(cromosoma):
  cromosoma = list(cromosoma)
  #print("Cromosoma original:", cromosoma)
  point1 = random.randint(1, len(cromosoma))
  cromosoma[point1:]=reversed(cromosoma[point1:])
  #print("Cromosoma después de la inversión:", cromosoma)
  return ''.join(cromosoma)

#print(inversionMutation(generate_cromosoma(cromosomaSize)))
```

```python
def genetic_algoritm():
  #Genero poblacion con unas list compreheition
  population = [generate_cromosoma(cromosomaSize) for _ in range(population_size)]
  print(population)
  for generation in range(generations): # Recorro las generaciones
    padre1,padre2=selection(population)
    #padre1,padre2=tournament_selection(population,3)
    print(f"Generacion {generation}: {padre1} (Fitness: {fitness(padre1)}) ")
    if fitness(padre1) <= 16:
      print (f"Solucion encontrada en la generacion {generation}!")
      break
    next_generation = [padre1,padre2]
    while len(next_generation) < population_size:
      #cruce=crossover(padre1, padre2)
      cruce2=crossover2(padre1, padre2)
      #child = mutation(cruce)
      #child = inversionMutation(cruce)
      child = mutation(cruce2)
      #child = inversionMutation(cruce2)
      next_generation.append(child)
    population = next_generation
```
